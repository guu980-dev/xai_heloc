```python
# Set plot style for better visuals
plt.style.use('ggplot')
sns.set(style="whitegrid")
```

```python
# Load dataset (assuming it's a CSV)
# Replace 'your_data.csv' with your actual file path
df = pd.read_csv('data/heloc_dataset_v1 (1).csv')
```

```python
# Check the column names
print("\nColumn Names:", df.columns.tolist())
```

# 2. Dealing with Missing Data

```python
# Check percentage of missing data per column
missing_percentage = df.isnull().sum() / len(df) * 100
print("\nMissing Data Percentage per Column:")
print(missing_percentage[missing_percentage > 0])
```

# 3. Univariate Analysis

```python
# Plot distribution of numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()

for column in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column].dropna(), kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Plot distribution of categorical columns
categorical_columns = df.select_dtypes(include=[object]).columns.tolist()

for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=df[column], order=df[column].value_counts().index)
    plt.title(f'Count of categories in {column}')
    plt.show()
```

# 4. Bivariate Analysis (Numerical Columns)

```python
# Correlation matrix for numerical variables
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
```

```python
# Heatmap for correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

# 6. Multivariate Analysis (Categorical vs Categorical)

```python
# Crosstab analysis
for category1 in categorical_columns:
    for category2 in categorical_columns:
        if category1 != category2:
            print(f'\nCrosstab between {category1} and {category2}')
            print(pd.crosstab(df[category1], df[category2]))
```

```python
# Heatmap for categorical vs categorical relationships
for category1 in categorical_columns:
    for category2 in categorical_columns:
        if category1 != category2:
            plt.figure(figsize=(10, 6))
            crosstab_data = pd.crosstab(df[category1], df[category2])
            sns.heatmap(crosstab_data, cmap="YlGnBu", annot=True, fmt="d")
            plt.title(f'Heatmap of {category1} vs {category2}')
            plt.show()
```

# 7. Outlier Detection (Numerical Columns)

```python
# Boxplots to detect outliers in numerical columns
for column in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(df[column])
    plt.title(f'Outliers in {column}')
    plt.show()
```

# 8. Data Skewness

```python
# Skewness of numerical features
skewness = df[numerical_columns].skew()
print("\nSkewness of Numerical Features:")
print(skewness)
```
